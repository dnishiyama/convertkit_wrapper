```python
# HOW MANY PEOPLE HAVE BOUGHT STUDIOMATES AND THEN CANCELLED?

# Find the email address of purchasers
monthlies = stripe_data[stripe_data['plan_name (metadata)']=='Monthly']; len(monthlies) # 77
annuals = stripe_data[stripe_data['plan_name (metadata)']=='Annual']; len(annuals) # 16
combined = list(set(list(annuals['Customer Email']) + list(monthlies['Customer Email']))); combined
purchasers = combined

# Find the current status of each of those purchasers
data={p:{} for p in purchasers}
for purchaser in purchasers:
    for idx, row_data in stripe_data[stripe_data['Customer Email'] == purchaser].iterrows():
        if data.get(purchaser, {}).get('state', 'canceled') != 'active':
            data[purchaser]['state'] = row_data['Status']
    rows = stripe_data[(stripe_data['Customer Email'] == purchaser) & (stripe_data['plan_name (metadata)'].isin(['Monthly','Annual']))]; rows
    stripe_data[stripe_data['Customer Email'] == purchaser]
    for s,e in zip(list(rows['Created (UTC)']), list(rows['Ended At (UTC)'])):
    #     interval = datetime.now() - datetime.strptime(s, '%Y-%m-%d %H:%M')
        if data[purchaser]['state'] != 'active':
            interval = datetime.strptime(e, '%Y-%m-%d %H:%M') - datetime.strptime(s, '%Y-%m-%d %H:%M')
            data[purchaser]['subscription_length'] = data[purchaser].get('subscription_length', 0) + interval.days
        data[purchaser]['start_date'] = datetime.strptime(s, '%Y-%m-%d %H:%M')
```
